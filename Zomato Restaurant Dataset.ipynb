```python
df['rate'].unique()
```

# Removing 'NEW','-' and '/5' from rate columns

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
    
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#But now as we have converted strings and other to 'nan'. So,it contains many 'null' values.
```

```python
# Number of rate rows containing 'null' values:
df['rate'].isnull().sum()
```

```python
# So, 10019 rows have 'null' values.
```

# Filling all the Null values in 'rate' column with mean.

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
#inplace=true implies that we are changing in the original array and not creating any duplicate array.
```

```python
df['rate'].isnull().sum()
```

```python
# Now, no null values exist.
```

```python
# We see that values have commas which rstricts it from becoming an integer so, we will first remove commas.
```

# Removing commas from Cost2Plates column

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates']=df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

# Analysing 'rest_type' now

```python
df['rest_type'].value_counts()
```

```python
# So, we see some restaurants type have counts in 10,000 , 20,000 etc but some have counts lesser than 5 or 10 so, we cluster
# all those restaurant types that have lesser counts.
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)  #printng in descending order
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Storing rest types lesser than 1000 in frequency under a common name called 'others'

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value;

df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
# All those rest_types which had frequeny lesser than 1000 are now stored under the name 'others'.
```

```python
df.head()
```

# Cleaning Location columns

```python
df['location'].value_counts()
```

```python
# Here also, we will make clusters of all those locations which are lesser than a particular value in frequency and group
# then under a common name.
```

```python
location= df['location'].value_counts(ascending=False)

location_lessthan300=location[location<300]
def handle_location(value):
    if(value in location_lessthan300):
        return 'others';
    else:
        return value;
    
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

# Cleaning 'cuisine' column:

```python
df['cuisines'].value_counts(ascending=False)
```

```python
# Again clustering all those cuisines which have lesser than 100 frequency i.e. value counts
cuisines=df['cuisines'].value_counts(ascending=False)

cuisines_lessthan100=cuisines[cuisines<100]
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

# Checking if 'Type' column is clean:

```python
df['Type'].value_counts(ascending=False)
```
